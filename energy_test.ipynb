```python

x = np.random.normal(-1, 1, size=(3,))
y = 1j*np.random.normal(-1, 1, size=(3,))

x = x.astype(np.clongdouble)
y = y.astype(np.clongdouble)
initial_state = x + y
initial_state = initial_state/np.linalg.norm(initial_state)


initial_state = np.array([1+1j,0,1])

initial_state = initial_state/scipy.linalg.norm(initial_state)


D0 = scipy.outer(scipy.conj(initial_state),initial_state)


s = 1
d = 1



H0 = scipy.array([[s,-d/3,-d/3],
               [-d/3,s,-d/3],
               [-d/3,-d/3,s]], dtype=np.clongdouble)


H0 = scipy.array([[1,0,2],
               [0,3,0],
               [2,0,1]], dtype=np.clongdouble)


eig, U = scipy.linalg.eig(H0)
Ut = scipy.linalg.inv(U)

for i in range(3):
    D0[i,i] = np.real(D0[i,i])


D0_E = Ut @ D0 @ U
```

```python
s0_E =  Ut @ initial_state
D = scipy.outer(scipy.conj(s0_E), s0_E)
```
