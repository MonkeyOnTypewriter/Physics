```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 100

x = np.linspace(-1,1,num=N)
dx = x[1]-x[0]
```

```python
H = (-1/(2*dx**2))*(-2*np.diag(np.ones(N)) + np.diag(np.ones(N-1), 1) + np.diag(np.ones(N-1), -1))
e, v = np.linalg.eigh(H)
v = v.T
```

```python
dt = .001
time = 10
steps = int(time/dt)

B = np.eye(N) - 1j*dt/2*H
A = np.linalg.inv(np.eye(N) + 1j*dt/2*H)

data = np.zeros([steps, N], dtype=np.complex128)
data[0] = v[0] + v[2]

for i, state in enumerate(data[1:]):
    data[i+1] = A@B@data[i]

```

```python
k = 0

plt.plot(x, data[k].real)
plt.plot(x, data[k].imag)
plt.ylim(-1,1)
```
