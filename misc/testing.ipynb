```python
import numpy as np
import Perturbations as pb
```

```python
# set initial hamiltonian
s = 1
d = .5
H = np.array([[s,-d/3,-d/3],
              [-d/3,s,-d/3],
              [-d/3,-d/3,s]])

# set initial density matrix
state0 = np.random.uniform(-1, 1, size=(3,)) + 1j*np.random.normal(-1, 1, size=(3,))
state0 = state0/np.linalg.norm(state0)
X = np.outer(np.conj(state0),state0)
```

```python
P = pb.get_pert_dist(X, H, type='diag', basis='position')
```

```python
pb.graph_pert_dist(P,X)
```

```python
d = pb.get_pert()

P = pb.calculate_pert_traj(X, H, d)

pb.graph_traj(X,P)
```

```python
ds = []
for i in range(100):
    d = pb.get_pert()
    ds.append(d)
```

```python
times = np.linspace(0, 100, num = 1000)
```

```python
from matplotlib import pyplot as plt
```

```python
lam = .1
for time in times:
    P = np.array([X.flatten()])
    for d in ds:
        pert = pb.calculate_pert_evol(X, H, d, l=lam, t=time)
        P = np.concatenate((P,pert))

    data = P
    x=[]
    y=[]
    for k in range(9):
        real = [z.real for z in data[:,k]]
        imag = [z.imag for z in data[:,k]]
        x.append(real)
        y.append(imag)

    Xf = X.flatten()

    re = [z.real for z in Xf]
    im = [z.imag for z in Xf]


    samp_mean = np.mean(data, axis=0)

    lw = .275

    indices = np.arange(len(data))

    figure, axis = plt.subplots(3, 3, figsize=(9.12,9.12))
    figure.suptitle(f'lambda = {lam}, time={round(time, 1)}')




    axis[0,0].scatter(x[0],y[0], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[0,0].plot(re[0],im[0], 'ro')

    center_point = [re[0],im[0]]
    axis[0,0].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[0,0].set_ylim(center_point[1] - .5, center_point[1] + .5)



    axis[0,1].scatter(x[1],y[1], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[0,1].plot(re[1],im[1], 'ro')



    center_point = [0,0]
    axis[0,1].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[0,1].set_ylim(center_point[1] - .5, center_point[1] + .5)



    axis[0,2].scatter(x[2],y[2], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[0,2].plot(re[2],im[2], 'ro')


    center_point = [0,0]
    axis[0,2].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[0,2].set_ylim(center_point[1] - .5, center_point[1] + .5)



    axis[1,0].scatter(x[3],y[3], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[1,0].plot(re[3],im[3], 'ro')

 
    center_point = [0,0]
    axis[1,0].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[1,0].set_ylim(center_point[1] - .5, center_point[1] + .5)


    axis[1,1].scatter(x[4],y[4], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[1,1].plot(re[4],im[4], 'ro')

    center_point = [re[4],im[4]]
    axis[1,1].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[1,1].set_ylim(center_point[1] - .5, center_point[1] + .5)

    axis[1,2].scatter(x[5],y[5], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[1,2].plot(re[5],im[5], 'ro')


    center_point = [0,0]
    axis[1,2].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[1,2].set_ylim(center_point[1] - .5, center_point[1] + .5)


    axis[2,0].scatter(x[6],y[6], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[2,0].plot(re[6],im[6], 'ro')


    center_point = [0,0]
    axis[2,0].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[2,0].set_ylim(center_point[1] - .5, center_point[1] + .5)


    axis[2,1].scatter(x[7],y[7], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[2,1].plot(re[7],im[7], 'ro')



    center_point = [0,0]
    axis[2,1].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[2,1].set_ylim(center_point[1] - .5, center_point[1] + .5)


    axis[2,2].scatter(x[8],y[8], linewidth=lw, alpha=1, c=indices, cmap='viridis')
    axis[2,2].plot(re[8],im[8], 'ro')

    center_point = [re[8],im[8]]
    axis[2,2].set_xlim(center_point[0] - .5, center_point[0] + .5)
    axis[2,2].set_ylim(center_point[1] - .5, center_point[1] + .5)


    re = [z.real for z in samp_mean]
    im = [z.imag for z in samp_mean]
    axis[0,0].plot(re[0],im[0], 'mX')
    axis[0,1].plot(re[1],im[1], 'mX')
    axis[0,2].plot(re[2],im[2], 'mX')
    axis[1,0].plot(re[3],im[3], 'mX')
    axis[1,1].plot(re[4],im[4], 'mX')
    axis[1,2].plot(re[5],im[5], 'mX')
    axis[2,0].plot(re[6],im[6], 'mX')
    axis[2,1].plot(re[7],im[7], 'mX')
    axis[2,2].plot(re[8],im[8], 'mX')

    plt.savefig(f'modl/{time}.png')
    plt.close()
```
